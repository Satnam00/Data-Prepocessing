```python
# pandas boxplot
df_gdp_2007.boxplot(by='continent', 
                    column=['lifeExp'], 
                    figsize=(12,6),
                    grid=True)
plt.title("Boxplot of Life Expectancy - by Continents for year 2007")
plt.suptitle("")
```

```python
# USING SEABORN plot library
```

```python
# Construct swarmplot

# Initialize Figure and Axes object
fig, ax = plt.subplots(figsize=(12,6))
ax.set_title("Swarmplot of Life Expectancy - by Continents for year 2007")
#plt.title("Swarmplot of Life Expectancy - by Continents for year 2007")
#plt.suptitle("")

sns.swarmplot(x="continent", y="lifeExp", data=df_gdp_2007, ax=ax)
#sns.swarmplot(x="continent", y="lifeExp", data=df_gdp_2007)
```

```python
# To make basic boxplot with Seaborn, we can use the pandas dataframe as input and use Seaborns 
# boxplot function. In addition to the data, we can also specify multiple options to customize the 
# boxplot with Seaborn. 
# Here, we have chosen colorblind friendly palette colorblind. Other color palette options 
# available in Seaborn include deep, muted, bright, pastel, and dark. Let us also specify the 
# width of the boxes in boxplot.
```

```python
# We will first use Seaborns boxplot like before with no data points and 
# add a layer of data points to the boxplot with stripplot. 

# Initialize Figure and Axes object
fig, ax = plt.subplots(figsize=(12,6))
ax.set_title("Boxplot of Life Expectancy - by Continents for year 2007")
sns.boxplot(x='continent',
            y='lifeExp', 
            data=df_gdp_2007, 
            width=.6,
            palette="pastel", 
            ax=ax
            )
sns.stripplot(x='continent',
            y='lifeExp', 
            data=df_gdp_2007, 
            jitter=True,
            palette="pastel", color='black',
            ax=ax
            )
```

```python
# Boxplot with Swarm plot using Seaborn
# Another way we can visualize data points with Seaborn boxplot is to 
# add swarmplot instead of stripplot. 
# Initialize Figure and Axes object
fig, ax = plt.subplots(figsize=(12,6))
ax.set_title("Boxplot of Life Expectancy - by Continents for year 2007")
plot=sns.boxplot(x='continent',
            y='lifeExp', 
            data=df_gdp_2007, 
            width=.6,
            palette="pastel", 
            ax=ax
            )
plot=sns.swarmplot(x='continent',
            y='lifeExp', 
            data=df_gdp_2007, 
            alpha=.5,
            color='black',
            ax=ax
            )
```

```python
# output file name
plot_file_name="test.jpg"
 
# save as jpeg
plot.figure.savefig(plot_file_name,
                    format='jpeg',
                    dpi=100)
```
