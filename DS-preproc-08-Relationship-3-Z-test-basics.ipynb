# Example - 1

```python
A= np.array([434, 694, 457, 534, 720,
             400, 484, 478, 610, 641,
             425, 636, 454, 514, 563,
             370, 499, 640, 501, 625,
             612, 471, 598, 509, 531])

mu    = 500
sigma = 100
n     = A.shape[0]
```

```python
sample_mu = A.mean()
sample_mu
```

#### Qs

The question is, are these studentsâ€™ SAT scores (M = 536) significantly greater than a population mean of 500 with a population standard deviation of 100? 

Note that the the maker of the SAT prep program claims that it will increase (and not decrease) your SAT score.  So, you would be justified in conducting a one-directional test. (alpha = .05).

** Answer **

The goal of the analysis is to test for a significant deviation between the sample mean and the population mean. In this case, the population standard deviation is a known parameter.

** one-sample Z test of the mean: **

$$ \LARGE Z = \frac {\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}}$$

- The numerator of the formula is equal to sample mean minus the population value. 
- The denominator is equal to standard error: the standard deviation of the population divided by the square root of n (the number of scores in the sample).

H0 : Means are same. No effect because of prep program

** Compute test statistic** - You will find that the sample mean is equal to 536.00. Now, compute your Z statistic:

```python
Z = ( sample_mu - mu)/(sigma/np.sqrt(n))
Z
```

** Conduct hypothesis test. **

Since the company promoting the SAT prep program specifically claims that the program will improve SAT scores, we will conduct a directional (i.e., one-tailed) test with alpha = .05. 



For a Z = 1.8, the probability is = .96407

and the significance level is .95

OR the other way around is 

$\alpha = .05$ the Z score is 1.6

Hence the H0 is rejected. Meaning there is an effect of prep program
